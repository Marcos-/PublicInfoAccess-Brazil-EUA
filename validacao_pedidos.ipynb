````python
import pandas as pd
````

````python
df = pd.read_csv('2018.csv')
````

````python
df.head()
````

````python
df = df.loc[(df["predict_prob"] > 0.40)]
````

````python
len(df)
````

````python
def categorize_text(text, llm):
    # response = llm.invoke('Can you add an appropriate category next to each of the following texts? Respond with one of the two categories: "related to diversity and inclusion" or "not related to diversity and inclusion". Answear only with the category.  \n\n' + text)
    prompt = """Can you respond an appropriate category to the following input. Respond with one of the following categories: racial-related or not-racial-related. Your answear should have less than 3 words.
            For example,
                Necessito dos dados sobre o número de estudantes negros matriculados em todos os cursos da área de saúde na Universidade Federal da Paraíba. Se for possível nos últimos três anos também.;07/06/2017;Não;Não;Pelo sistema (com avisos por email);Internet;2785337;Educação Superior; ; ;21/06/2017;Prezado Senhor, Segue, em anexo, resposta da Pró-Reitoria de Graduação.;Acesso Concedido;Resposta solicitada inserida no Fala.Br - racial-related;
                Como saber quais faculdade com curso de pós graduação em Educação Especial na área de deficiencia intelectual são reconhecidas pelo MEC e CEE? - not-racial-related
            Input:
        """
    response = llm.invoke(prompt + text)

    return response
````

````python
df2 = pd.DataFrame()
````

````python
df['validation'] = df["DetalhamentoSolicitacao"].apply(lambda x: categorize_text(x, llm))
````

````python
df.to_csv("llama_2020.csv")
````

````python
import re
````

````python
new_column = []
for values in df['validation']:
    if re.search(r'Not-racial-related', values, re.IGNORECASE):
        new_column.append(0)
    elif re.search(r'Racial-related', values, re.IGNORECASE):
        new_column.append(1)
    else:
        new_column.append(0)

df['result'] = new_column
````
